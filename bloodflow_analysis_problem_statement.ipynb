Problem Statement::
Bayesian Estimation & fluctuation Detection in Cerebral Blood Flow 


key Objective:
To estimate hidden physiological parameters — artery radius (r) and blood viscosity (μ) — from observed cerebral blood flow (Q_obs) using Bayesian inference and Poiseuille’s law, and to detect flow fluctuations that may signal hidden pathological conditions.

Process :
1.Model blood flow through a cerebral artery using Poiseuille’s equation.
2.Estimate posterior distributions of radius and viscosity using PyMC and noisy flow observations.
3.Incorporate a fluctuation detection mechanism in the model to:
    a.Compare estimated flow with expected normal flow range.
    b.Identify and label abnormal flow states: vasoconstriction, vasodilation, or blockage.

4.Visualize posterior distributions and generate alerts based on deviations.
